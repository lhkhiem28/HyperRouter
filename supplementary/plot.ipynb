```python
import glob
from natsort import natsorted
import numpy as np
import tqdm
from scipy.stats import entropy
import matplotlib.pyplot as plt
```

```python
n_layers = 1
n_samples = 1000
```

```python
gate_scores = natsorted(glob.glob("../ckps/pretraining/SMoE-S-enwik8/SMoE-S/gate_scores/*.npy"))
len(gate_scores)

for g in range(n_layers):
    scores = np.load(gate_scores[g])[:8, :]
    plt.figure(figsize=(24, 2))
    for i in range(scores.shape[0]):
        plt.subplot(1, scores.shape[0], i + 1)
        plt.bar(list(range(1, 17)), scores[i])
        plt.xticks([]); plt.yticks([])
```
